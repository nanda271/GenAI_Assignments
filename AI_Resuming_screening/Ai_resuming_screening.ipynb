```python
from transformers import pipeline
```

```python
# Load lightweight model (fast + stable)
generator = pipeline(
    "text-generation",
    model="google/flan-t5-small",
    max_length=512
)
```

```python
def run_pipeline(resume, job_description):

    # STEP 1: Extract
    extract_prompt = f"""
Extract skills, experience, and tools from this resume:

{resume}
"""
    extracted = generator(extract_prompt)[0]["generated_text"]

    # STEP 2: Match
    match_prompt = f"""
Compare resume with job description.

Resume:
{extracted}

Job Description:
{job_description}
"""
    matched = generator(match_prompt)[0]["generated_text"]

    # STEP 3: Score
    score_prompt = f"""
Give a score from 0 to 100 based on match.

Data:
{matched}
Return only a number.
"""
    score = generator(score_prompt)[0]["generated_text"]

    # STEP 4: Explain
    explain_prompt = f"""
Explain why this score was given.

Score: {score}
Details: {matched}
"""
    explanation = generator(explain_prompt)[0]["generated_text"]

    return {
        "Extracted": extracted,
        "Matched": matched,
        "Score": score,
        "Explanation": explanation
    }
```

```python
job_description = """
Data Scientist Role
Skills: Python, Machine Learning, NLP, SQL
Tools: TensorFlow, PyTorch
"""

strong_resume = """
Python, Machine Learning, NLP, SQL, Deep Learning
Worked with TensorFlow and PyTorch
3 years experience
"""

average_resume = """
Python, SQL
Basic Machine Learning
1 year experience
"""

weak_resume = """
Excel, Communication
No programming experience
"""
```

```python
print("===== STRONG =====")
print(run_pipeline(strong_resume, job_description))

print("\n===== AVERAGE =====")
print(run_pipeline(average_resume, job_description))

print("\n===== WEAK =====")
print(run_pipeline(weak_resume, job_description))
```
